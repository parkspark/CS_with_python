# CS - 16. Databases - 16.1 SQL vs NoSQL databases

Databases
💡A database is a collection of useful data of one or more related organizations structured in a way to make data an asset to the organization. A database management system is a software designed to assist in maintaining and extracting large collections of data in a timely fashion.

 

데이터베이스는 데이터를 조직의 자산으로 만들기 위한 방식으로 구조화된 하나 이상의 관련 조직의 유용한 데이터 모음입니다. 데이터베이스 관리 시스템은 대규모 데이터 모음을 적시에 유지 관리하고 추출하는 데 도움이 되도록 설계된 소프트웨어입니다.

 

SQL vs NoSQL databases
💡SQL stands for Structured Query Language. It’s used for relational databases. A SQL database is a collection of tables that stores a specific set of structured data. Some examples are PostgreSQL, MySQL, MariaDB etc.

NoSQL stands for Not Only SQL. It’s used for non-relational databases. A NoSQL database is a collection of collections that stores a specific set of unstructured data. Some examples are MongoDB, CouchDB, Redis etc.

 

SQL은 구조화된 쿼리 언어의 약자입니다. 관계형 데이터베이스에 사용됩니다. SQL 데이터베이스는 특정 구조화된 데이터 집합을 저장하는 테이블 모음입니다. 몇 가지 예로는 PostgreSQL, MySQL, MariaDB 등이 있습니다.
NoSQL은 Not Only SQL의 약자입니다. 비관계형 데이터베이스에 사용됩니다. NoSQL 데이터베이스는 특정 비정형 데이터 집합을 저장하는 컬렉션의 모음입니다. 몇 가지 예로는 MongoDB, CouchDB, Redis 등이 있습니다.

 

### SQL 데이터베이스
SQL(Structured Query Language) 데이터베이스는 관계형 데이터베이스(RDBMS)로 알려져 있습니다. 이는 데이터를 열과 행으로 구성된 테이블에 저장하는 구조를 가지고 있습니다. 각 테이블은 고유한 키를 통해 연결될 수 있으며, 이를 통해 복잡한 쿼리를 실행할 수 있습니다.

장점:

표준화된 언어(SQL)를 사용하여 쿼리를 작성하고 관리할 수 있습니다.
데이터의 무결성을 보장합니다. 즉, 데이터가 정확하고 일관성이 있는지 확인하는 기능을 제공합니다.
트랜잭션을 지원합니다. 이는 여러 작업을 한 단위로 묶어 처리하는 것을 말합니다.


단점:
수평 확장이 어렵습니다. 즉, 새로운 서버를 추가하여 데이터베이스의 용량을 증가시키는 것이 복잡할 수 있습니다.
복잡한 쿼리는 시스템을 느리게 만들 수 있습니다.

### NoSQL 데이터베이스
NoSQL(Not Only SQL) 데이터베이스는 비관계형 데이터베이스를 말합니다. 이는 SQL 데이터베이스와 달리 고정된 스키마가 없으며, 데이터는 JSON, 키-값 쌍 등 다양한 형태로 저장될 수 있습니다.

장점:

수평 확장이 쉽습니다. 즉, 새로운 서버를 추가하여 데이터베이스의 용량을 쉽게 증가시킬 수 있습니다.
유연한 데이터 모델을 가지고 있어, 변경 사항을 쉽게 처리할 수 있습니다.


단점:
표준화된 쿼리 언어가 없으므로, 각 NoSQL 데이터베이스마다 쿼리를 작성하는 방법이 다릅니다.
데이터의 일관성을 보장하는 것이 SQL 데이터베이스보다 어렵습니다.

둘을 사용할 결정할 때 중요하게 고려해야 할점은 

1. 데이터 무결성

2. 복잡한 쿼리 수행

3. 데이터의 정형화 여부

4. 데이터 구조

5. 대량의 데이터

6. 빠른 응답 시간 

정도로 볼수 있습니다.

 

1. 데이터 무결성이 중요한 경우 - SQL 이 트랜잭션 지원과 무결성 보장기능을 지원합니다.

주로 금융, 의료 같은 무결성이 중요한 환경에서 고려됩니다.

 

2. 복잡한 쿼리 

-  SQL의 관게형 모델이 DB의 복잡한 쿼리, JOIN연산을 처리하기 쉽습니다.

 

3. 데이터의 정형화 여부

- SQL 은 고정된 스키마를 가지므로 데이터의 구조가 미리 정해져 있는 경우에 좋습니다.

 

4. 대량의 데이터

- NoSQL은 수평 확장이 쉽기 때문에, 빅데이터를 다룰때 유리합니다.

 

5. 데이터 구조

- 3번과 비슷한 영역인데 데이터 구조가 자주 변경되는경우 NoSQL은 구조가 자주 변경되더라도 유연하게 대처할수있습니다.

 

6. 빠른 응답 시간

- 빠른 응답시간이 필요한 경우 NoSQL은 DB에 간단한 쿼리에 대해 빠른 응답시간을 가집니다.

실시간 어플리케이션, 캐싱, 캐시, 세션등 관리에 유리합니다.
