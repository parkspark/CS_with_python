## In Pyhton


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
# ë…¸ë“œë¥¼ í‘œí˜„í•˜ëŠ” í´ë˜ìŠ¤ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        if not self.head:
            self.head = Node(data)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(data)

    def display(self):
        elements = []
        current_node = self.head
        while current_node:
            elements.append(current_node.data)
            current_node = current_node.next
        print(elements)
# ê·¸ í›„, Linked listë¥¼ í‘œí˜„í•˜ëŠ” í´ë˜ìŠ¤ë¥¼ ìƒì„±í•œë‹¤.
```

```python
linked_list = LinkedList()
linked_list.append('A')
linked_list.append('B')
linked_list.append('C')
linked_list.display()  # ì¶œë ¥: ['A', 'B', 'C']

# ì´ì™€ ê°™ì´ íŒŒì´ì¬ì—ì„œ Linked listë¥¼ êµ¬í˜„í•  ìˆ˜ ìˆë‹¤.
```

```python
# linked list êµ¬í˜„
class Node:
    def __init__(self, x):
        self.val = x
        self.next = None

def has_cycle(head):
    if head is None:
        return False

    slow = head  # ê±°ë¶ì´
    fast = head  # í† ë¼

    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:  # ê±°ë¶ì´ì™€ í† ë¼ê°€ ë§Œë‚˜ë©´ ìˆœí™˜ ìˆë‹¤ê³  íŒë‹¨
            return True

    return False
```
