```python
# ì „ìœ„ ìˆœíšŒ
def pre_order(node):
    if node is not None:
        print(node.data, end=' ')
        pre_order(node.left)
        pre_order(node.right)

# ì¤‘ìœ„ ìˆœíšŒ
def in_order(node):
    if node is not None:
        in_order(node.left)
        print(node.data, end=' ')
        in_order(node.right)

# í›„ìœ„ ìˆœíšŒ
def post_order(node):
    if node is not None:
        post_order(node.left)
        post_order(node.right)
        print(node.data, end=' ')
```
