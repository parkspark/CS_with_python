```python
class DirectedGraph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node, neighbour):   
        if node not in self.graph:
            self.graph[node] = [neighbour]
        else:
            self.graph[node].append(neighbour)
```

DFSëŠ” í˜„ì¬ ë…¸ë“œì—ì„œ ê°€ëŠ¥í•œ ê¹Šê²Œ íƒìƒ‰í•˜ê³ , ë” ì´ìƒ ë°©ë¬¸í•  ë…¸ë“œê°€ ì—†ì„ ë•Œ ì´ì „ ë…¸ë“œë¡œ ëŒì•„ê°€ëŠ” ë°©ì‹ìœ¼ë¡œ ì‘ë™í•©ë‹ˆë‹¤.

(ì¶”í›„ì— DFS, BFSë¥¼ ë‹¤ë£¹ë‹ˆë‹¤)

```python
class DirectedGraph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, node, neighbour):   
        if node not in self.graph:
            self.graph[node] = [neighbour]
        else:
            self.graph[node].append(neighbour)

    def dfs(self, node, visited): #ì¶”ê°€
        if node not in visited:
            visited.append(node)
            if node in self.graph:
                for neighbour in self.graph[node]:
                    self.dfs(neighbour, visited)
        return visited
```

```python
g = DirectedGraph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('B', 'D')
g.add_edge('C', 'D')

print(g.dfs('A', []))  # ['A', 'B', 'D', 'C']
```
